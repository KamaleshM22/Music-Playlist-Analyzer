```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

1. Load data

```python
df = pd.read_csv("music_playlist.csv")
```

```python
print("Dataset:")
print(df)
```

2. Average duration

```python
avg_duration = df["Duration (sec)"].mean()
```

```python
print("\nAverage Duration of Songs:", avg_duration)
```

3. Most common genre

```python
most_common_genre = df["Genre"].mode()[0]
```

```python
print("\nMost Common Genre:", most_common_genre)
```

4. Bar chart - Duration per song

```python
plt.figure(figsize=(10, 5))
plt.bar(df["Song Name"], df["Duration (sec)"])
plt.xticks(rotation=45)
plt.title("Duration of Each Song")
plt.xlabel("Song")
plt.ylabel("Duration (sec)")
plt.tight_layout()
plt.show()
```

5A. Bar chart - Genre distribution

```python
genre_counts = df["Genre"].value_counts()
plt.figure(figsize=(7, 5))
plt.bar(genre_counts.index, genre_counts.values)
plt.title("Genre Distribution (Bar Chart)")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

5B. Pie chart - Genre distribution

```python
plt.figure(figsize=(7, 7))
plt.pie(genre_counts.values, labels=genre_counts.index, autopct="%1.1f%%")
plt.title("Genre Distribution (Pie Chart)")
plt.show()
```
